### List the most frequent tokens

```python
from nltk.probability import FreqDist
#list of stopwords and filtering the copus of the punctuations and the stopwords
fdist = FreqDist(brown.words())
top_ten = fdist.most_common(10)
print(top_ten)
```

```python
# number of characters, words, sentences and vocabulary of each of the fileids present:
from nltk.corpus import gutenberg
for fileid in gutenberg.fileids():
        num_chars = len(gutenberg.raw(fileid)) 
        num_words = len(gutenberg.words(fileid))
        num_sents = len(gutenberg.sents(fileid))
        num_vocab = len(set(w.lower() for w in gutenberg.words(fileid)))
        print("Number of characters : " , round(num_chars/num_words), "Number of words: ", round(num_words/num_sents), "Number of sentences: ", round(num_words/num_vocab), " Title: " , fileid)
```

```python
# the sentences in a particular
macbeth_sentences = gutenberg.sents('shakespeare-macbeth.txt')
macbeth_sentences
```

### Inaugral Address Corpora - Documents consisting of the presendential speeches over the years

```python
# fileids and the years
from nltk.corpus import inaugural
inaugural.fileids()
```

```python
#plot for seeing the usage of words america and citizen over the years
cfd = nltk.ConditionalFreqDist(
    (target, fileid[:4])
    for fileid in inaugural.fileids()
    for w in inaugural.words(fileid)
    for target in ['america','citizen']
    if w.lower().startswith(target)
)
cfd.plot()
```
